```python
#Setting forth the purpose
def null_detail(df):
  func_df= pd.DataFrame(index= df.columns, columns=["datatype", "total_values", "na_values", "non_na_values", "%na_values", "unique_count"])
  func_df["datatype"]= [type(i) for i in list(df.columns)]
  func_df["total_values"]= [len(df[i]) for i in list(df.columns)]
  func_df["na_values"]= [df[i].isnull().sum() for i in list(df.columns)]
  func_df["non_na_values"]= [(~df[i].isnull()).sum() for i in list(df.columns)]
  func_df["%na_values"]= [df[i].isnull().mean()*100 for i in list(df.columns)]
  func_df["unique_count"]= [len(df[i].unique()) for i in list(df.columns)]
  return func_df
```

## **8. Are reservations canceled as a result of a long waiting period?**

 ## **10. What month has the highest ADR for hotels?** 



```python
# group hotel and arrival date month
booking_monthly=df_hotel.groupby(['arrival_date_month','hotel'])['adr'].mean().reset_index()

# list the months.
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# It will take values and the month list's order from the dataframe.
booking_monthly['arrival_date_month']=pd.Categorical(booking_monthly['arrival_date_month'],categories=months,ordered=True)

# Sort the arrival date month column.
booking_monthly=booking_monthly.sort_values('arrival_date_month')

booking_monthly
```

```python
# Let's ploting of the arrival_date_month, adr, and hotel.
plt.figure(figsize=(22,13))
sns.lineplot(x=booking_monthly['arrival_date_month'],y=booking_monthly['adr'],hue=booking_monthly['hotel'])

# set lables
plt.title('ADR of each month')
plt.xlabel('Months')
plt.ylabel('Average Daily Rate')
```

**Observation**

The ADR for Resrot Hotel is greater in the months of June, July, and August when compared to City Hotels. This summer, guests/clients could want to stay at resort hotels.

Due to the low average daily rate during these months, January, February, March, April, October, November, and December are the best months for travellers to resort or city hotels.
